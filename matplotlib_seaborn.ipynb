### Subplot

[Subplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html) permite crear gr√°ficos dentro de una gr√°fica. Esto lo hace a trav√©s de una matriz de gr√°ficos y se puede acceder a ellos a trav√©s de √≠ndices:

plt.subplot(row, column, index)

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 101)
y = x ** 2

#plt.figure(figsize=(5,5)) #(width, height)

# Grafico 1
plt.subplot(2, 2, 1)
plt.plot(x, y, 'b:') # PLot 1
plt.plot(y, x, 'r--') # Plot 2
# Grafico 2
plt.subplot(2, 2, 2)
plt.hist(y) # Plot 3
plt.axhline(10, color='springgreen', linestyle=':') # Plot 4
# Grafico 3
plt.subplot(2, 2, 3)
plt.bar(['F', 'M'], [16, 38], color=['pink', 'cyan']) # Plot 5
# Grafico 4
plt.subplot(2, 2, 4)
plt.pie([1, 7], labels=['b', 'o'])
plt.legend(loc='lower right')
plt.show()
````

### Metodo orientado a objetos

Tenemos distintas maneras de mostrar graficas o visualizaciones dentro de Matplotlib, una de ellas es **Pyplot** es sencillo, pero es dificil de customizar y dificil de llegar a mas de una figura o entrar a ciertos detalles. Por otro lado esta el **Orientado a objetos** es mas complejo y necesita mas de codigo, pero brinda una mayor personalizacion, nos permite crear distintos tipos de graficos y distintos tipos de distribucion en uno solo.

Pyplot:
- Rapido
- Facil
- Una sola figura

Object Oriented:
- Mayor personalizacion
- Mas amigables a multiples diagramas
- Mas codigo

Un objeto define una **Figure** (lienzo) el cual contendra graficas que las llamaremos **Axes**, estos Axes pueden tener diferentes ejes (Axis).

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 101)
y = x**2

fig = plt.figure() # lienzo de las graficas
# los axes son las graficas dentro del lienzo (figure)
axes1 = fig.add_axes([1, 1, 0.5, 0.4]) # [row, column, width, height]
axes2 = fig.add_axes([1.05, 1.15, 0.2, 0.2])
axes1.plot(x, y, 'mediumslateblue')
axes2.plot(y, x, 'cyan')
axes2.set_facecolor('#333')
plt.show()
````

### Subplots

Sirve para trabajar multiples graficas en un solo lienzo, solo que **Subplots** aprovecha las bondades de trabajar con el metodo orientado a objetos de Matplotlib.

````python
# Subplots
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 101)
y = np.sin(x)

plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.plot(x, y, 'mediumslateblue')
plt.show()
````

````python
import matplotlib.pyplot as plt

x = np.linspace(0,10,101)
y = np.cos(x)

fig, ax = plt.subplots(nrows=1, ncols=2)
# se pueden acceder a los axes como si de arrays se tratasen
ax[0].plot(x, y, 'b')
ax[1].plot(y, x, 'r')
plt.show()
````

````python
import matplotlib.pyplot as plt

x = np.linspace(0,10,101)
y = np.cos(x)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
# se pueden acceder a los axes por medio de la desestructuracion 
ax1.plot(x, y, 'b')
ax2.plot(y, x, 'r')
fig.tight_layout() #mejora la visualizaci√≥n de los ejes de cada gr√°fico
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10, 101)
y = np.sin(x)

plt.style.use('ggplot')
fig, axes = plt.subplots(nrows=2, ncols=3)
axes[0,0].plot(x,y, 'g')
axes[0,1].plot(y,x, 'orange')
axes[0,2].hist(x, color='b')
axes[1,0].pie(x)
axes[1,1].scatter(x,y)
axes[1,2].boxplot(x)
fig.tight_layout()
plt.show()
````

### Leyendas, etiquetas, titulos, tama√±o

Para darle contexto a nuestros graficos necesitamos usar titulos, leyendas, tama√±o, etiquetas, para que nuestra grafica tenga un contexto mas amplio.

````python
# utilizando subplots()
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = np.sin(x)

plt.style.use('ggplot')
fig, axes = plt.subplots(1,2, figsize=(7,4)) # figsize defe la dimension del lienzo
axes[0].plot(x,y, label='sin(x)') # definimos las etiquetas con label
axes[0].set_title('x vs y') # titulos
axes[0].set_xlabel('x') # titulo eje x
axes[0].set_ylabel('y') # titulo eje y
axes[0].legend(loc='lower left') # muestra las leyendas
axes[1].plot(y,x, 'b', label='x')
axes[1].set_title('y vs x')
axes[1].set_xlabel('y')
axes[1].set_ylabel('x')
axes[1].legend()
plt.show()
````

````python
# utilizando pyplot
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = np.cos(x)

plt.style.use('ggplot')
plt.plot(x, y, label='$cos(x)$') # los $ le dan una notacion matematica
plt.title('x vs y')
plt.xlabel('x')
plt.ylabel('y')
# con bbox podemos cambiar el tama√±o y ubicacion de la leyenda
plt.legend(loc='upper center', bbox_to_anchor=(1.1,1)) 
plt.show()
````

Otros valores de *loc* de **[legend()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html)**:

### Bar Plot

Bar plot o grafica de barras, es un tipo de grafica que nos permite graficar variables categoricas.

[bar()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html)

[barh()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html)

````python
# bar plot
import matplotlib.pyplot as plt
import numpy as np

country = ['Italy', 'Japan', 'Mexico', 'Germany', 'China']
population = [100, 300, 800, 400, 999]

plt.style.use('ggplot')
plt.figure(figsize=(6,4))
plt.bar(country, population, width=0.4,  color=['violet','mediumslateblue'])
plt.title('Poblacion')
plt.ylabel('poblacion en millones')
plt.xlabel('paises')
# para cambiar los valores del eje x (lista_indices, valores_nuevos)
plt.xticks(np.arange(5), ('Italia', 'Japon', 'Mexico', 'Alemania', 'China'), rotation=360-25)
plt.tight_layout()
plt.show()
````

````python
# horizontal bar plot
import matplotlib.pyplot as plt
import numpy as np

country = ['Italy', 'Japan', 'Mexico', 'Germany', 'China']
population = [100, 300, 800, 400, 999]

plt.style.use('ggplot')
plt.figure(figsize=(6,4))
plt.barh(country, population, height=0.4)
plt.title('Poblacion')
plt.xlabel('poblacion en millones')
plt.ylabel('paises')
plt.yticks(np.arange(5), ('Italia', 'Japon', 'Mexico', 'Alemania', 'China'))
plt.tight_layout()
plt.show()
````

### Crear otro tipo de graficas

````python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randint(1,50,100)
````

[Hist](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html)

Muestra la distribucion de los datos.

````python
plt.hist(data, bins=5) # bins son el numero de barras para los rangos de frecuencia
plt.show()
````

[Boxplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.boxplot.html#matplotlib.axes.Axes.boxplot)

Nos permite ver como es la distribucion de nuestros datos. 

Es util para identificar outliers (datos anomalos)

[Explicacion bloxplot](https://www.labxchange.org/library/items/lb:LabXchange:d8863c77:html:1)

````python
plt.boxplot(data, vert=False, patch_artist=True, notch=True, medianprops={'color': 'orange'})
plt.show()

# la linea azul representa la mediana o percentil 50
# los extremos de la caja abarcan el 50%
# cada linea fuera de la caja abarca un 25% cada una
````

[Scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.scatter.html#matplotlib.axes.Axes.scatter)

O grafico de dispersion, nos sirve para observar la relacion entre dos variables, correlacion y su comportamiento.

````python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
colors = np.random.rand(N)

plt.style.use('ggplot')
plt.scatter(x, y, s=area, c=colors, alpha=0.6)
plt.colorbar()
# s: cambia el tama√±o de cada punto en referencia a una tercera variable
# c: cambia el color de cada punto en referencia a una cuarta variable
plt.show()
````

### Set

Nos permite configurar el estilo del fondo, la paleta de colores, fuente, y tama√±o de la fuente.

- Este par√°metro afecta a todas las gr√°ficas que creamos, incluidas las que no usan Seaborn.
- Podemos resetear los valores utilizando el m√©todo **reset_orig()**.
- Este par√°metro es un atajo de [set_theme](https://seaborn.pydata.org/generated/seaborn.set_theme.html#seaborn.set_theme), que es el recomendado seg√∫n la documentaci√≥n.
- Otro metodo para cambiar la paleta de colores es [color_palette](https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette).

````python
import seaborn as sns
import matplotlib.pyplot as plt

sns.reset_orig()
sns.barplot(x=['A','B','C'], y=[1,3,2], hue=['A','B','C'])
plt.show()
````

````python

import seaborn as sns
import matplotlib.pyplot as plt

sns.reset_orig()
#sns.set(style='darkgrid', palette='flare', font="", font_scale=1)
sns.barplot(x=['A','B','C'], y=[1,3,2], hue=['A','B','C'], palette='flare')
plt.show()
````

### Distribuciones

A continuacion se muestra el tipo de visualizaciones para valores numericos en temas de [ditribucion](https://seaborn.pydata.org/tutorial/distributions.html).

````python
import seaborn as sns

sns.set_style('darkgrid')
tips_dataset = sns.load_dataset('tips')
tips_dataset.head()
````

````python
# dist o diagramas de distribucion
# kind: cambiamos el tipo de grafica de distribucion (hist, kde, ecdf)
sns.displot(tips_dataset, x='tip', hue='sex', stat='count', kind='ecdf');
````

### Joinplot y Pairplot

Son dos tipos de graficos que son muy utiles en el mundo de la ciencia de datos porque con un unico comando podemos combinar mas de un tipo de grafico.

Son utiles para trabajar con datasets de entrada.    

[Joinplot](https://seaborn.pydata.org/generated/seaborn.jointplot.html) une dos distintas graficas en uno solo para poder sacar mas valor analitico.

[Pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html) muestra las distintas relaciones entre las variables numericas, eligiendo los graficos que seaborn cree mas adecuados.

````python
import seaborn as sns

tips_dataset = sns.load_dataset('tips')

# dispersion de los datos de total_bill vs tip
# distribucion de cada eje para ver la concentracion de datos
sns.jointplot(tips_dataset, x='total_bill', y='tip');
````

````python
sns.jointplot(tips_dataset, x='total_bill', y='tip', hue='sex');
````

````python
# podemos cambiar el tipo de visualizacion con kind
sns.jointplot(tips_dataset, x='total_bill', y='tip', hue='sex', kind='hist');
````

````python
# marginal_ticks: muestra la escala de las graficas externas
# marginal_kws: cambia la configuracion de las graficas externas
sns.jointplot(tips_dataset, x='total_bill', y='tip', 
            hue='sex', 
            kind='hist', 
            marginal_ticks=True,
            marginal_kws={'bins': 25, 'fill': True, 'multiple': 'stack'}
);
````

````python
# pairplot
sns.pairplot(tips_dataset);
````

````python
# corner: elimina las graficas que presenten informacion duplicada
sns.pairplot(tips_dataset, hue='sex', palette='flare', corner=True);
````

### Heatmap

[Heatmap](https://seaborn.pydata.org/generated/seaborn.heatmap.html) es un tipo de grafico muy util para graficar datos en una estructura matricial.

````python
import seaborn as sns

tips_dataset = sns.load_dataset('tips')

# llevamos el dataset a una estructura matricial
# cor() encuentra la correlacion entre las variables
tips = tips_dataset.corr(numeric_only=True)

sns.heatmap(tips);
````

````python
# annot: muestra los valores de la correlacion
# cmap: nos permite cambiar el mapa de colores
# linewidth: da una separacion entre los cuadros
sns.heatmap(tips_dataset.corr(numeric_only=True), 
            annot=True, 
            cmap='crest', 
            linewidth=3,
            linecolor='#333',
);
````

Nota: Un tip para cuando se quiera invertir los colores en cmap, es solo agregar ‚Äú_r‚Äù al final del nombre del color usado en cmap.

````python
# vmin y vmax nos permiten definir los valores minimo y maximo de la correlacion
sns.heatmap(tips_dataset.corr(numeric_only=True), 
            annot=True, 
            linewidth=2,
            vmin=0,
            vmax=1
);
````
